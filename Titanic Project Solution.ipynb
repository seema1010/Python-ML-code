# Step 2: Exploratory Data Analysis (EDA)

```python
# import required libraries for visualization
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
# First explore data for passengers age distribution
plt.figure(figsize=(10,8))# # note here 8 is no of passengers ie y label and 10 is age x label.
train_df['Age'].hist(bins = 70)  # total no of records for age col is 700 so 700 divided by 10 is 70
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.show()
```

```python
# Explore data for passengers survival by age. we are identifining if there is any correlation
# between age and survived. this may tell us that after age of 65 people not survived.
#here we use scatter plot and alpha= .55 is used to adjust transperency level and is between 0 
# to 1. This is related to graph view in laptop.
plt.figure(figsize=(10,14))  #note 14 is age ie y label and 10 is x label
plt.scatter(train_df.Survived, train_df.Age, alpha=0.55) 
plt.ylabel('Age')
# sets the y -axis lable
plt.ylabel('Age')
# Formats the grid style of our graphs
plt.grid(b=True, which='major', axis='y')
plt.title('Survival by Age (1=Survived)')
plt.show()
```

```python
# Explore data for passengers Survival Breakdown. This tells that more people died than no of
# people survived ie see graph.This is confirmed basis graph coming. note: survived is a column
# name in training data set.
plt.figure(figsize=(8,6)) # 8 is x label and 6 is y label coming in veritical side
train_df.Survived.value_counts().plot(kind='barh', color= 'blue', alpha=0.65)
plt.title('Survival Breakdown (1=Survived, 0 =Died)')
plt.show()
```

```python
# Explore data for passengers class distribution. . Note: Pclass is a column in training data
#set. Here graph confirms that 3rd class people died more.
plt.figure(figsize=(10,8))
train_df.Pclass.value_counts().plot(kind='barh', color= 'g', alpha=0.65)
plt.title('Class Distribution')
plt.show()
```

```python
# Explore data for passengers embarkment by Location.note: Embarked is column name in test data 
# setgraph shows passenger location and in which location more passenger died can be known.
plt.figure(figsize=(8,6))
train_df.Embarked.value_counts().plot(kind='bar', alpha=0.5)
plt.title('Passengers Embarkment by Location')
plt.show()
```

```python
# note in logistic regression if x is continous and y is categorical then we use logistic regression
#algo so here y is survived is response variable or traget variable we need to predict this and 
# so sex column x we need to convert to 0 0r 1 note this is like one hot encoding but this method 
# is different from one hot encoding.
# here we are doing pre - processing 
titanic_train_file=pd.read_csv("C:\\Users\\hp11\\Desktop\\machine learning training somen choudhary\\project work after day8 titanic\\train.csv")
cols=["Pclass","Sex","Age"]# here we take only 3 columns and we can take more but we need to 
#transform them 
#change male to 1 and female to 0 by using lambda function. it says if male put 1 and vice versa.
titanic_train_file["Sex"]=titanic_train_file["Sex"].apply(lambda sex:1 if sex=="male" else 0)
```

```python
#handle missing values of age. here we use below method
titanic_train_file["Age"]=titanic_train_file["Age"].fillna(titanic_train_file["Age"].mean())
titanic_train_file["Age"]=titanic_train_file["Fare"].fillna(titanic_train_file["Fare"].mean())
survived=titanic_train_file["Survived"].values
data=titanic_train_file[cols].values # here we create a new data file with the name as data.
# the transformed values will be stored in this file.
#Note: vase impute or after handling missing values distribution should not change. To check 
#this we can do before and after analysis.  For above code check create historgram and see if 
#there is a major shift then mean is not good.
```
